```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api.formatters import TextFormatter
from youtube_transcript_api import TranscriptsDisabled


video_ids = "Cu6KTdoOw2A,SWAiC9kHqKc,v0CkCWvggXw".split(',')

formatter = TextFormatter()

for video_id in video_ids:
    try:
        # Check if the video has a transcript in English or translate from Hindi
        transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)

        # Try to get Hindi transcript and translate to English
        transcript_hi = transcript_list.find_transcript(['hi'])
        if transcript_hi:
            translated_transcript = transcript_hi.translate('en')
            translated_data = translated_transcript.fetch()

            # Format and print the transcript
            text = formatter.format_transcript(translated_data)
            print(f"TRANSCRIPT FOR VIDEO ID {video_id}:\n\n{text}\n")
        else:
            print(f"NO HINDI TRANSCRIPT FOUND TO TRANSLATE FOR VIDEO ID: {video_id}")

    except TranscriptsDisabled:
        print(f"Subtitles/transcripts are disabled for video ID: {video_id}")
```
